```python
from bs4 import BeautifulSoup
import requests  #kunai pani url ma scrapping ko lagi request pathaucha
url = BeautifulSoup('https://www.worldometers.info/coronavirus/', 'html.parser')
soup = requests.get(url)
#soup.text   #yo garda url ko sabai source code string format ma aaucha
# so yo string lai xml or lxml ma convert garnu parcha which is the browser readable format
# beautifulsoup ma string related document ma kaam hudaina
soup = BeautifulSoup(soup.text, "lxml") #xml le chai data dherai cha bhaney dhanna sakdaina
soup
```

```python
table_code = soup.table # tag ko naam table bhako bhayera table liyeko
table_code # aba table ko code matra lincha
```

```python
tags = table_code.find_all('tr')
tags
```

```python
data = []
for tag in tags:
    y = tag.text.split('\n')  #tag.text le chai html content lidaina
    #tag bhitra bhako data matra aaucha
    #text chai hamle html content hatayera data matra lina ko lagi use gareko
    if y[1] != "":
        data.append(y[1:])

data

#.split() le string lai \n ko basis ma list ma change gardincha
```

```python
import csv
file = open('covid_data.csv','w')
x = csv.writer(file)
x.writerows(data)
file.close()
```

```python
import pandas as pd
df = pd.read_csv('covid_data.csv',encoding = 'latin1')
df
```

```python
countries = list(df['Country,Other'])
total_cases = df['TotalCases']
total_cases = list(map(lambda x:int(x.replace(',' , '')),list(df['TotalCases'] )))
```
